```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Import the unified dataframe all_data.csv
df_all = pd.read_csv('/datasets/gdrive/DS4A - Team Project/data/all_data_enhanced.csv', sep=',', parse_dates=['FECHA'], encoding="UTF-8")
```
